# Q1. Decision Frameworks

假設在時間 $t$，投資人先觀察歷史市場資訊 $X_t$，接著選擇投資組合配置 $A_t=a$。投資人做決策時，未來市場狀態 $S_{t+1}$ 與未來報酬 $R_{t+1}$ 未知。

股票池來自台股大型權值股，依照產業別大致分類後，選取各產業具代表性的標的：

| 產業 | 股票 |
|---|---|
| 半導體 | 台積電 2330、聯發科 2454 |
| 電子組件與散熱 | 台達電 2308、台光電 2383 |
| 電腦及週邊設備 | 鴻海 2317、廣達 2382 |
| 金融保險業 | 富邦金 2881、國泰金 2882 |
| 網通、電信與傳統產業 | 智邦 2345、中華電 2412 |


## Q1-1. State Space

State 表示未來市場狀態。令：

$$
S_{t+1} \in \mathcal S
$$

將市場狀態簡化為三類：

$$
\mathcal S = \{\text{Bull}, \text{Bear}, \text{Stress}\}
$$

| State | 意義 |
|---|---|
| $\text{Bull}$ | 市場趨勢良好，報酬為正，風險相對穩定 |
| $\text{Bear}$ | 市場趨勢轉弱，報酬偏負 |
| $\text{Stress}$ | 市場處於高波動、大幅回撤或極端負報酬狀態 |

這裡的 $S_{t+1}$ 是投資人做決策後所面對的未來市場狀態，不是投資人可以控制的變數。


## Q1-2. Data Generating Process

假設實際發生的時間順序如下：

1. 時間 $t$：投資人觀察歷史資訊 $X_t$。
2. 時間 $t$：投資人選擇投資組合配置 $A_t=a$。
3. 時間 $t+1$：未來市場狀態 $S_{t+1}$ 發生。
4. 時間 $t+1$：投資組合報酬 $R_{t+1}$ 實現。

因此，可將未來市場狀態的生成寫成：

$$
S_{t+1}\mid X_t \sim f_{S\mid X}(\cdot\mid X_t)
$$

投資人在時間 $t$ 可觀察的資訊包含近期報酬、波動、回撤與產業報酬：

$$
X_t = \{\text{Recent Return}, \text{Volatility}, \text{Drawdown}, \text{Sector Return}\}
$$

對於任一投資組合配置 $a$，未來報酬的條件分布可寫為：

$$
R_{t+1}\mid A_t=a, S_{t+1}=s \sim F_{a,s}
$$

其中：

$$
F_{a,s}(r)
=
P(R_{t+1}\leq r\mid A_t=a, S_{t+1}=s)
$$

也就是說，在不同未來市場狀態與不同投資組合配置下，未來報酬會有不同的條件分布。


## Q1-3. Action Space

Action 是投資人可以控制的投資組合配置。令：

$$
A_t \in \mathcal A
$$

設定三種候選 action：

$$
\mathcal A = \{a_1, a_2, a_3\}
$$

| Action | 投資組合配置 | 說明 |
|---|---|---|
| $a_1$ | Benchmark Portfolio | 依標準化後的市值權重配置 |
| $a_2$ | Semiconductor-Based Portfolio | 加碼台積電與聯發科 |
| $a_3$ | Defensive Portfolio | 加碼富邦金、國泰金 |

這三個 action 分別代表市場型配置、成長型配置與防禦型配置。


## Q1-4. Observation and Decision Rule

投資人在做決策前可以觀察到歷史資訊：

$$
X_t
$$

例如：

$$
X_t = \{\text{Recent Return}, \text{Volatility}, \text{Drawdown}, \text{Sector Return}\}
$$

Decision rule 是將觀察到的市場資訊轉換成投資組合配置的規則：

$$
d:X_t \mapsto A_t
$$

其中 $X_t$ 是投資人在時間 $t$ 可觀察到的資訊，$A_t$ 是最後選擇的投資組合配置。

因此可以理解為：投資人根據 $X_t$，透過 decision rule $d$，從 action space $\mathcal A$ 中選擇一個配置。


## Q1-5. Loss Function

給定市場狀態 $s$ 與投資組合配置 $a$，定義 loss function：

$$
\ell(s,a) = -\bar R_{a,s} + \lambda \sigma_{a,s} + \gamma |DD_{a,s}|
$$

其中：

| 符號 | 意義 |
|---|---|
| $\bar R_{a,s}$ | 投資組合 $a$ 在 state $s$ 下的平均報酬 |
| $\sigma_{a,s}$ | 投資組合 $a$ 在 state $s$ 下的波動度 |
| $DD_{a,s}$ | 投資組合 $a$ 在 state $s$ 下的回撤 |
| $\lambda$ | 波動風險懲罰參數 |
| $\gamma$ | 回撤風險懲罰參數 |

報酬越高，loss 越低，波動、回撤越大，loss 越高。

## Q1-6. Bayesian Decision Framework

在 Bayesian framework 中，未來市場狀態 $S_{t+1}$ 被視為隨機變數。投資人會根據目前資訊 $X_t$ 更新對未來市場狀態的 posterior probability：

$$
P(S_{t+1}=s\mid X_t)
$$


對任一投資組合配置 $a$，Bayesian posterior risk 定義為：

$$
\rho(a\mid X_t)
=
E[\ell(S_{t+1},a)\mid X_t]
$$

也就是：

$$
\rho(a\mid X_t)
=
\sum_{s\in\mathcal S}
\ell(s,a)P(S_{t+1}=s\mid X_t)
$$

Bayesian optimal action 為：

$$
a^{Bayes}
=
\arg\min_{a\in\mathcal A}\rho(a\mid X_t)
$$

因此，在 Bayesian framework 下，投資人會根據 $X_t$ 判斷 Bull、Bear、Stress 各自的 posterior probability，再選擇 posterior expected loss 最小的投資組合。


## Q1-7. Frequentist Decision Framework

在 Frequentist framework 中，不對 Bull、Bear、Stress 設定 prior probability，而是把每一個未來市場狀態固定住，分別評估每個 action 在各 state 下的表現。

給定未來 state $s$，action $a$ 的 frequentist risk 為：

$$
\mathfrak r(s,a)
=
E[\ell(S_{t+1},a)\mid S_{t+1}=s]
$$

因此可以建立 risk matrix：

| Action | Bull | Bear | Stress |
|---|---:|---:|---:|
| $a_1$ Benchmark | $\mathfrak r(\text{Bull},a_1)$ | $\mathfrak r(\text{Bear},a_1)$ | $\mathfrak r(\text{Stress},a_1)$ |
| $a_2$ Semiconductor | $\mathfrak r(\text{Bull},a_2)$ | $\mathfrak r(\text{Bear},a_2)$ | $\mathfrak r(\text{Stress},a_2)$ |
| $a_3$ Defensive | $\mathfrak r(\text{Bull},a_3)$ | $\mathfrak r(\text{Bear},a_3)$ | $\mathfrak r(\text{Stress},a_3)$ |

Frequentist optimal action 為：

$$
a^{Frequentist} = \arg\min_{a\in\mathcal A}\max_{s\in\mathcal S}\mathfrak r(s,a)
$$

Frequentist framework 是選擇特定的投資組合使得在市場狀態最糟糕的情況下，損失最小，因此比較保守。


## Q1-8. Interventional Decision Framework

在 Interventional framework 中，資產配置被視為一種主動介入。

interventional framework 是在看：

$$
P(R_{t+1} \mid do(A_t=a), see(X_t=x))
$$

也就是投資人主動將配置設定為 $a$ 後，未來報酬的分布。

## Q1-9. Interventional Risk

在 interventional framework 中，其 DGP 為：

$$
S_{t+1} \mid do(A_t=a), see(X_t=x)
\sim f^{do}(s \mid do(a), see(x))
$$

而三種 action 可以視為三種 intervention：

| Intervention | 意義 |
|---|---|
| $do(a_1)$ | 主動持有 benchmark portfolio |
| $do(a_2)$ | 主動加碼半導體股票 |
| $do(a_3)$ | 主動加碼防禦型股票 |

對任一投資組合 action $a$，未來報酬的 interventional distribution 可寫成：

$$
F^{do(a)}(r) = P(R_{t+1}\leq r \mid do(A_t=a), see(X_t=x))
$$

Interventional risk 定義為：

$$
\rho^{do}(a)=E[\ell(S_{t+1}^{do(a)},a)]
$$

Interventional optimal decision 為：

$$
a^{do}=\arg\min_{a\in\mathcal A}E[\ell(S_{t+1}^{do(a)},a)]
$$

## Q1-10. Decision Framework 比較

| Framework | State 的處理方式 | Risk 定義 | Optimal Action |
|---|---|---|---|
| Bayesian | 對未來市場狀態設定 posterior probability | Posterior expected loss | 選擇 posterior expected loss 最小的 action |
| Frequentist | 固定每個未來 state 分別評估 | State-specific risk | 選擇 worst-case risk 最小的 action |
| Interventional | 將 action 視為主動介入 | Expected loss under $do(a)$ | 選擇介入後 expected loss 最小的 action |

# Q2. Variance Estimation

令每日收盤價為 $P_t$，log return 定義為：

$$
r_t=\log\left(\frac{P_t}{P_{t-1}}\right)
$$

給定樣本：

$$
r_1,r_2,\ldots,r_n
$$

目標是估計報酬變異數：

$$
\operatorname{Var}(r_t)=\sigma^2
$$

以下分別用 Frequentist、Bayesian 與 M-estimation 三種角度說明。


## Q2-1. Frequentist Framework

在 Frequentist framework 中，$\sigma^2$ 為未知參數。用 sample variance 作為估計。

樣本平均為：

$$
\bar r=\frac{1}{n}\sum_{t=1}^n r_t
$$

樣本變異數為：

$$
s^2=\frac{1}{n-1}\sum_{t=1}^n(r_t-\bar r)^2
$$

因此在 Frequentist framework 下，變異數估計量為：

$$
\hat\sigma_F^2=s^2
$$

使用歷史樣本資料，直接估計未知的變異數。


## Q2-2. Bayesian Framework

在 Bayesian framework 中，變異數 $\sigma^2$ 被視為隨機變數。先設定 prior，再透過資料更新成 posterior。

假設報酬服從常態分布：

$$
r_t \sim  N(\mu, \sigma^2)
$$

假設 prior 服從 inverse-gamma 分布：

$$
\sigma^2 \sim IG(\alpha, \beta),
$$

其中， $IG(\alpha, \beta)$ 是 inverse-gamma 分布伴隨參數 $\alpha, \beta$，可以用以下方式更新參數：
\begin{align*}
\alpha &= \alpha_0 + \frac{n}{2} \\
\beta &= \beta_0 + \frac{1}{2}\sum_{i = 1}^n (r_i - \overline{r})^2. \\
\end{align*}

最後， posterior variance 為：
$$
\mathbb{E}[\sigma^2 \mid r_1, \ldots, r_n] = \frac{\beta}{\alpha - 1}
$$


## Q2-3. M-estimation / Robust Estimation

最小化：
$$
\sum_{i = 1}^n \rho(r_i - \mu)
$$

其中， $\rho(\cdot)$ 是 robust loss function ，(此處假設使用 Huber loss). 在估計出平均值 $\hat{\mu}$ 之後，變異數可透過以下方式估計：
$$
\hat{\sigma}^2 = \frac{1}{n} \sum_{i = 1}^n (r_i - \hat{\mu})^2.
$$

Huber loss 定義為
$$
\ell_{\delta}(a) = 
\begin{cases}
\frac{1}{2}a^2 \quad \text{for} \ \vert a \vert \leq \delta, \\ 
\delta \cdot (\vert a \vert - \frac{1}{2}\delta) \quad \text{otherwise}.
\end{cases}
$$


## Q2-4. 三種估計方法比較

| Framework | 對 $\sigma^2$ 的看法 | 估計方式 | 優點 | 限制 |
|---|---|---|---|---|
| Frequentist | 未知參數 | Sample variance | 簡單、直觀、常用 | 對 outliers 敏感 |
| Bayesian | 隨機變數 | Prior + likelihood 得到 posterior | 可使用 prior 資訊 | 需要設定 prior 與分布假設 |
| M-estimation / Robust | 透過 robust loss 定義估計量 | Huber loss| 對極端值較穩健 | 結果依賴 loss function 與 tuning parameter |


# Q3. Risk Decomposition

此處證明兩個 risk decomposition：

1. MSE decomposition：risk 可以分解成 bias squared 與 variance。
2. Excess risk decomposition：excess risk 可以分解成 approximation error 與 estimation error。


## Q3-1. MSE Decomposition

令 $\hat\theta$ 為參數 $\theta$ 的 estimator，並定義 squared-error risk：

$$
\mathfrak r(\hat\theta)
=
E[(\hat\theta-\theta)^2]
$$

Bias 定義為：

$$
\operatorname{bias}(\hat\theta)
=
E[\hat\theta]-\theta
$$

Variance 定義為：

$$
\operatorname{var}(\hat\theta)
=
E[(\hat\theta-E[\hat\theta])^2]
$$

證明：

$$
\mathfrak r(\hat\theta)
=
\operatorname{bias}^2(\hat\theta)+\operatorname{var}(\hat\theta)
$$

分解誤差：

$$
\hat\theta-\theta
=
(\hat\theta-E[\hat\theta])+(E[\hat\theta]-\theta)
$$

代入 squared-error risk：

$$
\mathfrak r(\hat\theta)
=
E\left[\left((\hat\theta-E[\hat\theta])+(E[\hat\theta]-\theta)\right)^2\right]
$$

展開：

$$
\begin{aligned}
\mathfrak r(\hat\theta)
&=
E[(\hat\theta-E[\hat\theta])^2]
+2(E[\hat\theta]-\theta)E[\hat\theta-E[\hat\theta]]
+(E[\hat\theta]-\theta)^2
\end{aligned}
$$

由於：

$$
E[\hat\theta-E[\hat\theta]]=0
$$

所以中間項為 0。因此：

$$
\mathfrak r(\hat\theta)
=
\operatorname{var}(\hat\theta)+\operatorname{bias}^2(\hat\theta)
$$


## Q3-2. Excess Risk Decomposition

令 $\varrho(f)$ 表示 predictor $f$ 的 population risk。

定義函數空間中的最佳函數為：

$$
f^*=
\arg\min_f \varrho(f)
$$

假設空間 $\mathcal H$ 中的最佳函數為：

$$
f^*_{\mathcal H}
=
\arg\min_{f\in\mathcal H}\varrho(f)
$$

令資料估計出來的模型為 $\hat f_{\mathcal H}$。

Excess risk 定義為：

$$
\varrho^{xs}(\hat f_{\mathcal H})
=
\varrho(\hat f_{\mathcal H})-\varrho(f^*)
$$

Approximation error 定義為：

$$
\varrho^{app}_{\mathcal H}
=
\varrho(f^*_{\mathcal H})-\varrho(f^*)
$$

Estimation error 定義為：

$$
\varrho^{est}(\hat f_{\mathcal H})
=
\varrho(\hat f_{\mathcal H})-\varrho(f^*_{\mathcal H})
$$

從 excess risk 開始：

$$
\varrho^{xs}(\hat f_{\mathcal H})
=
\varrho(\hat f_{\mathcal H})-\varrho(f^*)
$$

加減同一項 $\varrho(f^*_{\mathcal H})$：

$$
\begin{aligned}
\varrho^{xs}(\hat f_{\mathcal H})
&=
\varrho(\hat f_{\mathcal H})-\varrho(f^*_{\mathcal H})
+\varrho(f^*_{\mathcal H})-\varrho(f^*) \\
&=
\varrho^{est}(\hat f_{\mathcal H})
+
\varrho^{app}_{\mathcal H}
\end{aligned}
$$

因此：

$$
\varrho^{xs}(\hat f_{\mathcal H})
=
\varrho^{app}_{\mathcal H}
+
\varrho^{est}(\hat f_{\mathcal H})
$$


## Q3-3. Interpretation

MSE decomposition 說明 estimator 的誤差來自兩部分：

| 部分 | 意義 |
|---|---|
| Bias squared | 系統性偏誤 |
| Variance | estimator 本身的隨機波動 |

Excess risk decomposition 則說明模型誤差來自兩部分：

| 部分 | 意義 |
|---|---|
| Approximation error | 假設空間 $\mathcal H$ 不夠大或不夠正確所造成的誤差 |
| Estimation error | 因為使用有限樣本估計模型所造成的誤差 |


# Q4. Black-Litterman Model

Black-Litterman model 是一個 Bayesian 的 portfolio allocation framework。
  
核心想法是將市場均衡報酬作為 prior，再將投資人主觀 views 作為 evidence，最後得到 posterior expected returns。


## Q4-1. Bayesian Foundation：Prior、Likelihood、Posterior

Black-Litterman model 的 Bayesian 結構可以寫成三個部分。

### Prior：Market Equilibrium Return

市場均衡報酬通常記為：

$$
\Pi
$$

假設：

$$
\Pi=\delta \Sigma w_{mkt}
$$

其中：

| 符號 | 意義 |
|---|---|
| $\Pi$ | 市場隱含均衡報酬 |
| $\delta$ | 風險厭惡係數 |
| $\Sigma$ | 資產報酬 covariance matrix |
| $w_{mkt}$ | 市值權重投資組合 |

### Likelihood：Investor Views

投資人 views 通常寫成：

$$
P\mu=Q+\varepsilon
$$

其中：

$$
\varepsilon \sim N(0,\Omega)
$$

| 符號 | 意義 |
|---|---|
| $P$ | Pick matrix，用來指定 view 作用在哪些資產或投資組合 |
| $Q$ | View vector，代表投資人主觀看法 |
| $\Omega$ | View uncertainty matrix，代表 views 的不確定性 |

### Posterior：Updated Expected Returns

Black-Litterman 的目標是將 prior 與 views 結合，得到 posterior expected returns：

$$
\mu_{BL}
$$


## Q4-2. Black-Litterman Model

Black-Litterman model 假設資產報酬服從 multivariate normal distribution。Posterior mean 公式為：

$$
\mu_{BL}
=
\left[(\tau\Sigma)^{-1}+P^\top\Omega^{-1}P\right]^{-1}
\left[(\tau\Sigma)^{-1}\Pi+P^\top\Omega^{-1}Q\right]
$$

其中：

| 符號 | 意義 |
|---|---|
| $\Pi$ | 市場均衡報酬，也就是 prior mean |
| $\Sigma$ | 資產報酬 covariance matrix |
| $\tau$ | 控制 prior mean uncertainty 的參數 |
| $P$ | Pick matrix |
| $Q$ | View vector |
| $\Omega$ | View uncertainty matrix |

- 若 prior 比較可靠，posterior 會接近 $\Pi$。
- 若 views 比較可靠，posterior 會更接近 $Q$。

當 view uncertainty 很大時：

$$
\Omega \rightarrow \infty
$$

則：

$$
\Omega^{-1}\rightarrow 0
$$

此時 views 幾乎沒有影響，因此：

$$
\mu_{BL}\rightarrow \Pi
$$


## Q4-3. Posterior Covariance and Portfolio Decision

Black-Litterman 不只更新 expected return，也可以得到 posterior uncertainty。

Posterior covariance of the mean estimate 為：

$$
V_{\mu,post}
=
\left[(\tau\Sigma)^{-1}+P^\top\Omega^{-1}P\right]^{-1}
$$

其中， $V_{\mu,post}$ 代表 posterior mean estimate 的不確定性。

因此，Black-Litterman 的輸出不只是 posterior expected return $\mu_{BL}$，也包含 posterior uncertainty。


## Q4-4. Strict Bayesian Interpretation

Strict Bayesian interpretation 嘗試讓 Black-Litterman 的參數具有更清楚的統計意義：

$$
\mu \sim N(\Pi,\tau\Sigma)
$$

而 $\tau\Sigma$ 可以被理解為 prior mean 的 covariance ，因此 $\tau$ 控制的是對 prior mean $\Pi$ 的信心：

| $\tau$ | 意義 |
|---|---|
| 較小 | prior uncertainty 較低，posterior 更接近 market equilibrium |
| 較大 | prior uncertainty 較高，views 對 posterior 影響較大 |


## Q4-5. Idzorek Confidence-Based Model

Idzorek model 的主要貢獻是將投資人直覺上的 confidence level 轉換成 view uncertainty matrix $\Omega$。

Idzorek 方法讓投資人用 0% 到 100% 的信心水準表達 views 的可靠程度。

| Confidence | 對 posterior 的影響 |
|---|---|
| 0% | view 完全沒有影響，posterior 接近 market equilibrium |
| 100% | view 被完全採用，posterior 貼近 view |
| 介於 0% 與 100% | posterior 介於 market equilibrium 與 view 之間 |


## Q4-6. Entropy Pooling

Entropy Pooling 是 Meucci 提出的更一般化方法。它不一定要求資產報酬服從常態分布，而是直接調整整個 prior distribution。

其核心是最小化 posterior distribution 與 prior distribution 之間的 KL divergence：

$$
p^*=
\arg\min_p D_{KL}(p\Vert p_0)
$$

其中：

$$
D_{KL}(p\Vert p_0)
=
\sum_{j=1}^J p_j\log\left(\frac{p_j}{p_{0,j}}\right)
$$

限制條件通常包含：

$$
\sum_{j=1}^J p_j=1
$$

$$
p_j\geq 0
$$

以及 views 對應的 constraints，例如：

$$
Ap=b
$$

也就是在滿足投資人 views 的限制下，選擇一個與 prior 距離最小的 posterior distribution。


## Q4-7. Summary Comparison of Black-Litterman Variants

| Variant | 假設分布 | 輸入 | 核心重點 | 優點 | 限制 |
|---|---|---|---|---|---|
| Black-Litterman | Gaussian / Normal | $\tau$, $\Omega$, $P$, $Q$ | Matrix Bayesian update | 結構清楚、計算方便 | $\tau$ 與 $\Omega$ 設定較主觀 |
| Strict Bayesian | Gaussian / Normal | Prior uncertainty | Bayesian interpretation | 統計意義較明確 | 實作仍需要設定 prior uncertainty |
| Idzorek | Gaussian / Normal | 信心水準 0%~100% | 將信心轉換成 $\Omega$ | 非常直觀 | confidence 到 $\Omega$ 的轉換仍有假設 |
| Entropy Pooling | Any distribution | Prior distribution + constraints | 最小化 KL divergence | 可處理 fat tails 與非線性 views | 計算與設定較複雜 |
